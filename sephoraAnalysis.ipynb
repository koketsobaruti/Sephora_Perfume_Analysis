```python
#find out the top 20 most selling fragrances in womens category
womenfragrances_df = fragrance_df[fragrance_df['secondary_category']=='Women']
top_100_women_fragrances = womenfragrances_df.nlargest(100, 'loves_count')
top_100_women_fragrances
```

```python
#we need to normalize the loves_count and reviews column
#we have used min_max

#first get top 10 fragrances
top_10_women_fragrances = top_100_women_fragrances.head(10)
top_10_women_fragrances[['loves_count','reviews']] = top_10_women_fragrances[['loves_count','reviews']]/top_10_women_fragrances[['loves_count','reviews']].max()
top_10_women_fragrances
```
